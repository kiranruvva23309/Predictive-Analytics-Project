```python
import pandas as pd
df = pd.read_excel(r"C:\Users\ASUS\Downloads\UcdpPrioConflict_v25_1.xlsx")
df.shape
df.head()
df.info()
df.columns
```

```python
df.isnull().sum()
df["territory_name"] = df["territory_name"].fillna("Unknown")
df["ep_end_date"] = pd.to_datetime(df["ep_end_date"], errors="coerce")
df = df.drop_duplicates()
df.to_excel(r"C:\Users\ASUS\Downloads\cleaned_conflict_data.xlsx", index=False)
```
